```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_countries_by_credit_rating'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table1=soup.find('table',class_="sortable wikitable")
```

```python
table_list=soup.find_all('table')[0]
```

```python
print(table_list)
```

```python
title_list=soup.find_all('th')
```

```python
title_list
```

```python
final_list=[title.text for title in title_list]
```

```python
SnP=final_list[:5]
```

```python
S_and_P=pd.DataFrame(columns=SnP)
```

```python
S_and_P
```

```python
S_and_P_column_data=table_list.find_all('tr')
```

```python
S_and_P_column_data
```

```python
for row in S_and_P_column_data[1:]:
    row_data=row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]

    length=len(S_and_P)
    S_and_P.loc[length]=individual_row_data
#print(individual_row_data)
```

```python
res = [sub[3:] for sub in S_and_P['Rating']] 
S_and_P['Rating']=res
```
